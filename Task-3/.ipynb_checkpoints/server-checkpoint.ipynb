```python
import socket
import numpy
import cv2
import pickle
```

```python
port = 2222
ip = '127.0.0.1'
```

```python
with socket.socket(socket.AF_INET, socket.SOCK_DGRAM) as s:
    s.bind((ip, port))
    while True:
        conn = s.recvfrom(100000000)
        client_ip = conn[1][0]
        print(client_ip)
        data = conn[0]
        data = pickle.loads(data)
        data = cv2.imdecode(data, cv2.IMREAD_COLOR)
        cv2.imshow('VideoCAP', data)
        if cv2.waitKey(10) == 13:
            break
    cv2.destroyAllWindows()
```
